## Decision Tree classifier

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Read the iris dataset

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['species'].value_counts()
```

### Seperate X and Y(species)

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

### Preprocessing the data for X

```python
from sklearn.pipeline import Pipeline 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Perform train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Create a decsion tree model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=1, min_samples_split=3, min_samples_leaf=3, criterion='gini')
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Hyperparameter tuning 
Optimize the hyperparameters i.e. max_depth, min_samples_split, min_samples_leaf, criterion

```python
params = {'max_depth':[2, 3, 4, 5, 6, 7],
          'min_samples_split':[5, 6, 7, 8, 9],
          'min_samples_leaf':[5, 6, 7, 8, 9],
          'criterion':['gini', 'entropy']}
```

```python
from sklearn.model_selection import GridSearchCV
dtc = DecisionTreeClassifier()
gscv = GridSearchCV(dtc, param_grid=params, cv=5, scoring='f1_macro')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

### Check accuracy score in train and test

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest, ytest)
```

## Predict the results for train and test

```python
ypred_tr = best_dtc.predict(xtrain)
ypred_ts = best_dtc.predict(xtest)
```

```python
ypred_tr[0:5]
```

```python
ytrain.head()
```

```python
ypred_ts[0:5]
```

```python
ytest.head()
```

### Plot the confusion matrix for test data

```python
best_dtc.classes_
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_ts)
cfd = ConfusionMatrixDisplay(cf, display_labels=best_dtc.classes_)
cfd.plot()
```

### Print the classification report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_ts))
```

### Model evaluation complete

### Plotting the decision tree

```python
import matplotlib.pyplot as plt 
from sklearn.tree import plot_tree
plt.figure(figsize=(12, 6))
plot_tree(best_dtc, filled=True, feature_names=xtrain.columns, class_names=best_dtc.classes_)
plt.show()
```

### Predict out of sample data

```python
xnew = pd.read_csv('iris_sample.csv')
xnew
```

### Apply pre.transform

```python
num_pipe
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

### Predict the results with probability

```python
preds = best_dtc.predict(xnew_pre)
preds
```

```python
probs = best_dtc.predict_proba(xnew_pre)
probs
```

```python
import numpy as np
max_probs = np.max(probs, axis=1)
max_probs
```

### Store above results in xnew dataframe

```python
xnew['species_pred']=preds
xnew['prob'] = max_probs
```

### Save above to results.csv
